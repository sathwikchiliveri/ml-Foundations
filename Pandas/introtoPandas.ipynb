```python
data_set=pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')
print(data_set.head)
print(data_set.head())
```

```python
print(data_set.tail())
```

```python
print(data_set.describe())
```

```python
print(data_set.info())
```

```python
##Accessing specific columns
print(data_set['species'].head())
print(data_set['species'].value_counts())
print(data_set['species'].tail())
print(data_set['sepal_length'].head())
```

```python
#Accessing multiple columns
print(data_set[['sepal_length','sepal_width']].head())
```

```python
#Accesing rows using loc and iloc
print(data_set.iloc[0])
#print(data_set.iloc[148:150])
print(data_set.loc[0])
```

```python
#conditional filtering
print(data_set['sepal_length']>6)
print(data_set[data_set['sepal_length']>6])
```

```python
print(data_set[(data_set['sepal_length']>5) & (data_set['species']=='setosa')])
```

```python
#column operations
##Adding  column
data_set['sepal_ratio']=data_set['sepal_length']/data_set['sepal_width']
print(data_set.head())
```

```python
#Data cleaning and Aggregation
#check for null values 
print(data_set.isnull())
#we  cantlook at all the 150 rows and 5 columns so we can use any() function to check if there are any null values in the dataset
print(data_set.isnull().sum())
```

```python
#Aggregation with groupby
print(data_set.groupby('species').mean())
print(data_set.groupby('species')['sepal_length'].agg(['mean','min','max']))
print(data_set.groupby('species')['sepal_width'].agg(['mean','min','max','count']))
```

```python
#sorting operations
#print(data_set.sort_values(by='sepal_length', ascending=True))
print(data_set.sort_values(by='sepal_length', ascending=False))
```

```python
#merging and joining
df1=pd.DataFrame({'id':[1,2,3],'Name':['Alice','Bob','Charlie']})
df2=pd.DataFrame({'id':[1,2,3],'Age':[25,30,35]})
merged_df=pd.merge(df1,df2,on='id')
print(merged_df)
```
