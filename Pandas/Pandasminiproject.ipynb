```python
import pandas as pd
#Load and understand the dataset
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
titanic_data = pd.read_csv(url)
print(titanic_data.head())
```

```python
print(titanic_data.info())
print(titanic_data.describe())
```

```python
#checking  missing values
print(titanic_data.isnull().sum())
```

```python
#fill missing values in 'Age' and fill them column with mean age
titanic_data['Age'].fillna(titanic_data['Age'].mean(), inplace=True)
titanic_data.info()
```

```python
#Drop columns like 'Cabin' which have a lot of missing values and are not relevant for our analysis
titanic_data.drop(columns=['Cabin'], inplace=True)
titanic_data.info()
```

```python
#Analyze survival rates 
#what is the overall survival rate
titanic_data['Survived'].mean()*100
titanic_data.head()

```

```python
#survival rate by gender
titanic_data.groupby('Sex')['Survived'].mean()*100
```

```python
#survivalrate by passenger class
titanic_data.groupby('Pclass')['Survived'].mean()*100
```

```python
#mean age of survivors vs non survivors
titanic_data.groupby('Survived')['Age'].mean()
```

```python
#survivial rate of chilren<16 vs adults
titanic_data[(titanic_data['Survived'] == 1) & (titanic_data['Age'] < 16)]
```

```python
titanic_data[titanic_data['Age']<16]['Survived'].mean()*100
```

```python
titanic_data[titanic_data['Survived']==1].sort_values(by='Age', ascending=True)
```

```python
bins=[0,16,60, float('inf')]
labels=['Child','Adult','Senior']
#create the AgeGroup column
titanic_data['AgeGroup']=pd.cut(titanic_data['Age'], bins=bins, labels=labels,right=False)
titanic_data.head()
```

```python
titanic_data.groupby('AgeGroup')['Survived'].mean()*100
```
